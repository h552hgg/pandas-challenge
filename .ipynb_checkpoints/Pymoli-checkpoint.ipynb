```python
import pandas as pd 
#import csv
file1=('/Users/jonathanrocha/Desktop/pandas hw/HeroesOfPymoli/Resources/purchase_data.csv')
file_df=pd.read_csv(file1)
file_df.head() # verify file is recieved and the output of the file
```

```python
# Player Count and DataFrame
player_count=len(file_df)
player_count_df=pd.DataFrame({"Player Count": [player_count]})
player_count_df
```

```python
#Purchase Analysis breakdown
```

```python
Unique_items=len(file_df["Item Name"].unique())
#Unique_items
```

```python
Revenue=file_df["Price"].sum()
#Revenue
```

```python
#Purchasing Analysis Dataframe
Puranalysis=pd.DataFrame({"Number of Unique Items": [Unique_items], "Average Price": [avepurchaseprice],"Total Number of Purchases": [Totalpurchases],"Total Revenue":[Revenue]})
Puranalysis["Total Revenue"]=Puranalysis["Total Revenue"].map("${:.2f}".format)
Puranalysis["Average Price"]=Puranalysis["Average Price"].map("${:.2f}".format)
Puranalysis
```

```python
# Age Demographics
# Recheck of the origin file and verify the columns 
file1=file_df
#file1.head()
```

```python
# Creation of the bins and group names to setup for the grouping by Age Ranges 
bins=[0, 9.9, 14.9, 19.9, 24.9,29.9,34.9,39.9,99.9]
group_names=["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
```

```python
file1["Age Ranges"]= pd.cut(file_df["Age"],bins,labels=group_names, include_lowest=True)
#file1.head()
```

```python
agerange_group=file1.groupby("Age Ranges")
rangecount=agerange_group.count()
#rangecount
```

```python
rangeage=rangecount["SN"]
#rangeage
```

```python
rangeper=rangeage/780
#rangeper
```

```python
# Age Demographics DataFrame
Age_demo=pd.DataFrame({"Total Count":rangeage,"Percentage of Players":rangeper})
Age_demo["Percentage of Players"]=Age_demo["Percentage of Players"].map("{:,.2%}".format)

Age_demo
```

```python
# Created a groupby
itemgroup=file_df.groupby(["Item Name","Item ID"])
itemgroup.head() 
```

```python
itemcount=itemgroup["Purchase ID"].count()
#itemcount
```

```python
item=itemgroup["Price"].mean()
#item
```

```python
itemtotal=itemgroup["Price"].sum()
#itemtotal
```

```python
# Most Popular Items
Popular_items1=pd.DataFrame({"Purchase Count": itemcount,"Item Price":item,"Total Purchase Value":itemtotal})
Popular_items1["Item Price"]=Popular_items1["Item Price"].map("${:.2f}".format)
Popular_items1["Total Purchase Value"]=Popular_items1["Total Purchase Value"].map("${:.2f}".format)
Popular_items=Popular_items1.sort_values(by="Purchase Count", ascending=False)
Popular_items.reset_index().set_index(["Item ID","Item Name"]).head()
```

```python
Profitable1=pd.DataFrame({"Purchase Count": itemcount,"Item Price":item,"Total Purchase Value":itemtotal})
Profitable1["Item Price"]=Profitable1["Item Price"].map("${:.2f}".format)
Profitable1["Total Purchase Value"]=Profitable1["Total Purchase Value"].map("${:.2f}".format)
Profitable=Profitable1.sort_values(by="Purchase Count", ascending=False)
Profitable.reset_index().set_index(["Item ID","Item Name"]).head()
```
